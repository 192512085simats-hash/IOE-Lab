```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler

# Create sample dataset
data = {
    "Temperature": [28, 29, 30, 31, 29, 30, 32, 33, 28, 29],
    "Humidity": [65, 66, 68, 69, 67, 70, 72, 71, 66, 68],
    "Pressure": [1012, 1011, 1010, 1013, 1011, 1012, 1014, 1015, 1011, 1012],
    "HeartRate": [78, 80, 81, 79, 80, 82, 84, 83, 79, 81],
    "SpO2": [98, 99, 97, 98, 99, 98, 97, 99, 98, 98]
}

df = pd.DataFrame(data)

# Display original data
print("Original Dataset:")
print(df)

# Check missing values
print("\nMissing Values:")
print(df.isnull().sum())

# Remove duplicates
df = df.drop_duplicates()

# Normalize data
scaler = MinMaxScaler()
df[df.columns] = scaler.fit_transform(df)

print("\nNormalized Dataset:")
print(df)

# Plot Temperature
plt.plot(df["Temperature"], marker='o')
plt.title("Normalized Temperature")
plt.xlabel("Sample")
plt.ylabel("Temperature")
plt.grid(True)
plt.show()
```
