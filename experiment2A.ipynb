```python
import random
import time
import pandas as pd
import matplotlib.pyplot as plt
from IPython.display import clear_output

# Store data
temperature = []
humidity = []
status = []

print("Starting Real-Time IoT Stream Analytics...\n")

# Simulated Publisher & Subscriber
for i in range(20):

    # Publisher generates sensor data
    temp = round(random.uniform(10, 40), 2)
    hum = round(random.uniform(30, 90), 2)

    # Subscriber receives data
    temperature.append(temp)
    humidity.append(hum)

    # Analytics
    mean_temp = sum(temperature) / len(temperature)

    if len(temperature) >= 5:
        moving_avg = sum(temperature[-5:]) / 5
    else:
        moving_avg = mean_temp

    # Anomaly Detection
    if temp > 35 or temp < 15:
        result = "Anomaly"
    else:
        result = "Normal"

    status.append(result)

    # Display
    clear_output(wait=True)

    print("Reading:", i + 1)
    print("Temperature:", temp, "°C")
    print("Humidity:", hum, "%")
    print("Mean Temperature:", round(mean_temp, 2))
    print("Moving Average:", round(moving_avg, 2))
    print("Status:", result)

    # Graph
    plt.figure(figsize=(8,4))
    plt.plot(temperature, marker='o', label='Temperature')
    plt.axhline(35, color='red', linestyle='--', label='Upper Limit')
    plt.axhline(15, color='green', linestyle='--', label='Lower Limit')
    plt.xlabel("Sample")
    plt.ylabel("Temperature (°C)")
    plt.title("Real-Time Temperature Monitoring")
    plt.legend()
    plt.grid(True)
    plt.show()

    time.sleep(1)

# Save CSV
df = pd.DataFrame({
    "Temperature": temperature,
    "Humidity": humidity,
    "Status": status
})

df.to_csv("sensor_data.csv", index=False)

print("\nExperiment Completed Successfully!")
print("Data saved as sensor_data.csv")
```
