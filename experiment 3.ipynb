```python
import random
import pandas as pd
import matplotlib.pyplot as plt

num_slots = 100

slot_number = []
status = []

vacant_slots = []
occupied_slots = []

for i in range(1, num_slots + 1):

    # 70% Occupied, 30% Vacant
    if random.random() < 0.7:
        s = "Occupied"
        occupied_slots.append(i)
    else:
        s = "Vacant"
        vacant_slots.append(i)

    slot_number.append(i)
    status.append(s)

# Create DataFrame
df = pd.DataFrame({
    "Parking Slot": slot_number,
    "Status": status
})

print(df)

print("\n========== PARKING SUMMARY ==========")
print("Total Slots      :", num_slots)
print("Occupied Slots   :", len(occupied_slots))
print("Vacant Slots     :", len(vacant_slots))

print("\nVacant Slot Numbers:")
print(vacant_slots)

print("\nOccupied Slot Numbers:")
print(occupied_slots)

if len(occupied_slots) > 85:
    print("\nParking Status : HIGH TRAFFIC")
else:
    print("\nParking Status : NORMAL")

# Save CSV
df.to_csv("parking_slots.csv", index=False)

# Plot
plt.figure(figsize=(14,4))

colors = ['red' if x=="Occupied" else 'green' for x in status]

plt.bar(slot_number, [1]*num_slots, color=colors)

plt.xlabel("Parking Slot Number")
plt.title("Smart Parking Occupancy")
plt.yticks([])

plt.show()
```
