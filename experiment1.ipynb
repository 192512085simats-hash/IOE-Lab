```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler

# Sample IoE Sensor Dataset
data = {
    "Soil_Moisture": [45, 50, 55, 48, 52, 60, 58, 54, 49, 53],
    "Air_Temperature": [26, 27, 28, 29, 30, 31, 29, 28, 27, 30],
    "Humidity": [60, 62, 65, 64, 66, 68, 67, 65, 63, 66],
    "Light_Intensity": [400, 420, 450, 470, 460, 500, 490, 480, 430, 450],
    "Battery_Level": [95, 94, 93, 92, 91, 90, 89, 88, 87, 86]
}

# Create DataFrame
df = pd.DataFrame(data)

# Display Original Dataset
print("Original Dataset:")
print(df)

# Check Missing Values
print("\nMissing Values:")
print(df.isnull().sum())

# Remove Duplicate Records
df = df.drop_duplicates()

# Normalize Dataset
scaler = MinMaxScaler()
df[df.columns] = scaler.fit_transform(df)

# Display Normalized Dataset
print("\nNormalized Dataset:")
print(df)

# Plot Air Temperature
plt.figure(figsize=(8,5))
plt.plot(df["Air_Temperature"], marker='o')
plt.title("Normalized Air Temperature")
plt.xlabel("Sample Number")
plt.ylabel("Normalized Value")
plt.grid(True)
plt.show()
```
